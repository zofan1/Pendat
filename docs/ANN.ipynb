# ANN BACKPROGRATION

```python
import pandas as pd
import numpy as np
```

```python
dt= pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
dt
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
y = dt['species']
x = dt.drop(['species'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=27)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=1000, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.001)
```

```python
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```
