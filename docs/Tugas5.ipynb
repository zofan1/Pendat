# Implementasikan SklearnPCA Pada Data Iris

```python
import seaborn as sns
from sklearn.model_selection import train_test_split

iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2)    # 2. Instantiate the model with hyperparameters
model.fit(X_iris)              # 3. Fit to data. Notice y is not specified!
```

```python
print("check number of components", model.n_components_)
```

```python
X_2D = model.transform(X_iris) # 4. Transform the data to two dimenstions
```

```python
print("show first row of data", X_2D[0,:])
```

```python
# transform each row of data
print(model.transform(X_iris.iloc[0, :].values.reshape(1, -1)))
```

```python
iris['PCA1'] = X_2D[:,0]
iris['PCA2'] = X_2D[:,1]
```

```python
import seaborn as sns

sns.lmplot(x="PCA1",y="PCA2", hue='species', data=iris, fit_reg=False, height=8, scatter_kws={"s": 150})
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components=3)    # 2. Instantiate the model with hyperparameters
model.fit(X_iris)              # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(X_iris) # 4. Transform the data to two dimenstions

iris['PCA1'] = X_2D[:,0]
iris['PCA2'] = X_2D[:,1]
iris['PCA3'] = X_2D[:,2]
```

```python
iris.head()
```

```python
import seaborn as sns
sns.set()

sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False, height=6, scatter_kws={"s": 150})
sns.lmplot(x="PCA1", y="PCA3", hue='species', data=iris, fit_reg=False, height=6, scatter_kws={"s": 150})
sns.lmplot(x="PCA2", y="PCA3", hue='species', data=iris, fit_reg=False, height=6, scatter_kws={"s": 150})
```

```python
iris.head()
```

```python
from sklearn.model_selection import train_test_split

X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
Xtrain.head()
```

```python
ytrain.head()
```

```python
Xtrain = Xtrain[['PCA1', 'PCA2', 'PCA3']]

Xtrain.head()
```

```python
Xtest = Xtest[['PCA1','PCA2','PCA3']]

Xtest.head()
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class

model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
```

```python
y_model = model.predict(Xtest)  
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model)
```

```python
import matplotlib.pyplot as plt

sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, cmap='viridis',
            xticklabels=['setosa', 'versicolor', 'virginica'], yticklabels=['setosa', 'versicolor', 'virginica'])

plt.title('Confusion matrix');
plt.xlabel('true label');
plt.ylabel('predicted label');
plt.show()
```


## Menggunakan KNN

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_model)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(ytest, y_model))
cm 
```

## Menggunakan Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=21)
clf_gini.fit(Xtrain, ytrain)
```

```python
y_pred_gini = clf_gini.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score with criterion gini index: {0:0.2f}'. format(accuracy_score(ytest, y_pred_gini)))
```

```python
y_pred_train_gini = clf_gini.predict(Xtrain)
y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.2f}'. format(accuracy_score(ytrain, y_pred_train_gini)))
```

```python
print('Training set score: {:.2f}'.format(clf_gini.score(Xtrain, ytrain)))
print('Test set score: {:.2f}'.format(clf_gini.score(Xtest, ytest)))
```

```python
plt.figure(figsize=(10,6))

from sklearn import tree

tree.plot_tree(clf_gini.fit(Xtrain, ytrain))
plt.show()
```

## Menggunakan ANN

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=1000, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.001)
```

```python
clf.fit(Xtrain, ytrain)
y_pred=clf.predict(Xtest)
```

```python
accuracy_score(ytest, y_pred)
```
